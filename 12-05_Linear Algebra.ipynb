## Matrix Operations

### Matrix Addition

```python
# Define the two matrices
AA = [[1, 2, 3],
      [4, 5, 6]]

BB = [[9, 8, 7],
     [6, 5, 4]]

# Initialize an empty matrix for the result
result = [[0, 0, 0],
          [0, 0, 0]]

# Iterate through each element of the matrices and add them
for i in range(len(AA)): #no. of rows     #i = 0,1 ==> 0
    for j in range(len(AA[0])): #no.of columns # j =0,1,2 => (0,0), (0,1), (0,2) 
        result[i][j] = AA[i][j] + BB[i][j]
        
result
```

```python
for i in result:
    print(i)
```

## Matrix Subtraction

```python
# Define the two matrices
AA = [[1, 2, 3],
      [4, 5, 6]]

BB = [[9, 8, 7],
     [6, 5, 4]]

# Initialize an empty matrix for the result
result = [[0, 0, 0],
          [0, 0, 0]]

# Iterate through each element of the matrices and add them
for i in range(len(AA)): #no. of rows     #i = 0,1 ==> 0
    for j in range(len(AA[0])): #no.of columns # j =0,1,2 => (0,0), (0,1), (0,2) 
        print(i,j)
        result[i][j] = AA[i][j] - BB[i][j]
        
result
```

```python
for i in result:
    print(i)
```

### Matrix multiplication

While multiplying a matrix with other matrix, we must  figure out the dimensions of the resultant matrix. For this, the rows of the first matrix and the columns of the second matrix will be the dimensions of the resultant matrix.

```python
# Define the two matrices
matrix1 = [[1, 2],  #3x2
           [4, 5],
           [2, 5]]

matrix2 = [[9, 8, 7, 6],  #2x4
           [6, 5, 4, 7]]

result = np.dot(matrix1, matrix2)

print(result)
print("shape: ", result.shape)
```

## Transpose

```python
E = np.array([[1,2,3],[7,8,9]]) #2x3 ==> 3x2

transp = np.transpose(E)
print(transp)
print("\n")
print(E.shape)
print(transp.shape)
```

## Inverse of a Matrix

The formulae to calculate the inverse of a matrix is as follows:​

(A)-1 =  1/|A|. (Adj A)

The determinant of a matrix is the scalar value or number calculated using a square matrix. The square matrix could be 2×2, 3×3, 4×4, or any type.

```python
#inverse
F = np.array([[2,1],[4,3]]) # 1/|A|.(Adj A)
inverse = np.linalg.inv(F)
inverse
```
