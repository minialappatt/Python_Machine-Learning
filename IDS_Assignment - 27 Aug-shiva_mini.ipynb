## 3.1 Check for

* duplicate data

```python
##---------Type the code below this line------------------##

# Check for duplicate data 
duplicate_records = df[df['DR_NO'].duplicated()]
print("\n\033[1mDuplicate Rows in DR_NO: \033[0m")

# Print duplicate data if it exists, else print a message
if not duplicate_records.empty:
    cols_with_duplicates = duplicate_records.columns
    print("\n\033[1mDuplicate Data:\033[0m ")
    print(duplicate_records[cols_with_duplicates])
else:
    print("No duplicates found.")


# Check for missing data in DR_NO
missing_data = df.isnull().sum()

# Filter and print columns with missing data
columns_with_missing = missing_data[missing_data > 0]
print("\n\033[1mColumns with Missing Data: \033[0m")
print(columns_with_missing)

#Adding Sex code for future plotting
df['Vict Sex Cd'] = df['Vict Sex'].replace(['M','F','H','X',''],[0,1,2,3,9])

# Check for date inconsistencies
#check if date reported is before than the date occurred
df["DATE OCC"] = pd.to_datetime(df["DATE OCC"])
df["Date Rptd"] = pd.to_datetime(df["Date Rptd"])
date_time_inconsistencies = df[df["DATE OCC"] > df["Date Rptd"]]

# Check for time inconsistencies
invalid_time_occ = df[df['TIME OCC'] > 2359]

print("\n\033[1mDate and Time Inconsistencies:\033[0m")
if date_time_inconsistencies.empty:
    print("\n No date and time inconsistencies.")
else:
    print("\n\033[1m Date and Time Inconsistencies:\033[0m")
    print(date_time_inconsistencies)

    
if invalid_time_occ.empty:
    print("\n No invalid time references.")
else:
    print("\n\033[1m Invalid Time Occurrences:\033[0m")
    print(invalid_time_occ)

# Check for geographic inconsistencies/missing data
geographic_inconsistencies = df[
    (df["LAT"].isnull()) |
    (df["LON"].isnull()) |
    ((df["LAT"] == 0) & (df["LON"] == 0))
]

print("\n\033[1mGeographic Inconsistencies (LAT and LONG is 0 OR NULL):\033[0m")
geographic_inconsistencies

if not geographic_inconsistencies.empty:
    print(f"\n\033[1m Count of Geographic inconsistencies: {len(geographic_inconsistencies)}\033[0m\n")
    #print(geographic_inconsistencies)
    print(geographic_inconsistencies[["LAT", "LON"]])
else:
    print("\nNo Geographic inconsistencies found.")

# Check for age inconsistencies
age_inconsistencies = df[df["Vict Age"] <= 0]

print("\n\033[1mAge Inconsistencies (Negative or Zero Age of Victim):\033[0m")
if not age_inconsistencies.empty:
    print(f"\n\033[1m Count of age inconsistencies: {len(age_inconsistencies)}\033[0m\n")
    print(age_inconsistencies[["Vict Age"]])
    
else:
    print("\nNo age inconsistencies found.")

```

```python
##---------Type the code below this line------------------##

# Remove duplicate records based on 'DR_NO' column and keep the first occurrence
df = df.drop_duplicates(subset='DR_NO', keep='first')

# Impute missing values in 'Vict Sex' column with the most frequent value
most_common_sex = df['Vict Sex'].mode()[0]
df['Vict Sex'].fillna(most_common_sex, inplace=True)

most_common_descent = df['Vict Descent'].mode()[0]
df['Vict Descent'].fillna(most_common_descent, inplace=True)

df['Vict Sex Cd'] = df['Vict Sex'].replace(['M','F','H','X',''],[0,1,2,3,9])

# Impute 'Premis Desc' column with 'Not available' for NULL or blank values
df['Premis Desc'].fillna('Not available', inplace=True)
df['Premis Desc'] = df['Premis Desc'].apply(lambda x: 'Not available' if pd.isna(x) or x.strip() == '' else x)


# Check if columns are already dropped
columns_to_drop = ['Crm Cd 2', 'Crm Cd 3', 'Crm Cd 4', 'Cross Street']
dropped_columns = []


for column in columns_to_drop:
    if column in df.columns:
        df.drop(columns=[column], inplace=True)
        dropped_columns.append(column)

if dropped_columns:
    print("Dropped columns:", dropped_columns)
else:
    print("No columns were dropped.")


# Impute 'Weapon Used Cd' with 0 and 'Weapon Desc' with "No Weapon Used"
df['Weapon Used Cd'].fillna(0, inplace=True)
df['Weapon Desc'].fillna("No Weapon Used", inplace=True)


# Impute 'Mocodes' with "Unknown" (Modus Operandi Codes)
df['Mocodes'].fillna("Unknown", inplace=True)

# Print the count of missing values after imputation
remaining_missing_values = df.isnull().sum()
columns_missing_after_imputing1 = remaining_missing_values[remaining_missing_values > 0]


# Calculate median latitude and longitude for non-zero values
median_lat = df[df["LAT"] != 0]["LAT"].median()
median_lon = df[df["LON"] != 0]["LON"].median()

# Replace 0 latitude and longitude with median values
df.loc[df["LAT"] == 0, "LAT"] = median_lat
df.loc[df["LON"] == 0, "LON"] = median_lon

print("\n \033[1m Corrected LAT and LONG fields:\033[0m")
print(df[["LAT", "LON"]])

# Print the count of negative age values after removal
negative_age_count = (df['Vict Age'] <= 0).sum()


# Remove rows with unrealistic or negative Vict Age values
df = df[df['Vict Age'] > 0]

print(f" Number of negative and Zero (value) age values removed:\033[0m {negative_age_count}")


df['Vict Sex Cd'].fillna(9, inplace=True)


if not columns_missing_after_imputing1.empty:
    #columns_missing_after_imputing = columns_missing.columns
    print("\n \033[1m Missing Values After Imputation:\033[0m ")
    print(columns_missing_after_imputing1)#[columns_missing_after_imputing])
else:
    print("\n \033[1m No Missing Values After Imputation\033[0m ")
```

## 3.4 Text data

1. Remove special characters
2. Change the case (up-casing and down-casing).
3. Tokenization — process of discretizing words within a document.
4. Filter Stop Words.

```python
##---------Type the code below this line------------------##


# Remove special characters
df['Crm Cd Desc'] = df['Crm Cd Desc'].apply(lambda x: re.sub(r'\W+', ' ', x))


# Change the case (Camel Casing on col Crm Cd Desc)
df['Crm Cd Desc'] = df['Crm Cd Desc'].apply(lambda x: ' '.join([w.capitalize() for w in x.split()]))

# Tokenization
df['Crm Cd Desc Tokens'] = df['Crm Cd Desc'].apply(word_tokenize)

# Filter Stop Words
stop_words = set(stopwords.words('english'))
df['Tokens After Stop Words Filtering'] = df['Crm Cd Desc Tokens'].apply(lambda tokens: [word for word in tokens if word.lower() not in stop_words])

df.head()
```

```python
#Clustering using KMeans
#Justification: KMeans is a popular and simple clustering algorithm that can help group similar data points together.


from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Assuming your dataframe is named 'df'
X = df[['LAT', 'LON']]  # Consider using location attributes for clustering

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Determine the optimal number of clusters using the Elbow method
inertia_values = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia_values.append(kmeans.inertia_)

# Plot the Elbow curve
plt.plot(range(1, 11), inertia_values, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()

# Choose an optimal number of clusters based on the plot and fit KMeans
optimal_clusters = 3  # You should choose the optimal number from the plot
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

# Add cluster labels to the dataframe
df['Cluster'] = clusters
print(df[['LAT', 'LON', 'Cluster']])
```
