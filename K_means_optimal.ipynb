```python
import numpy as np
import pandas as pd
```

```python
df= pd.read_csv('customers.csv')
df.head()
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['Age'],df['Spending Score (1-100)'])
plt.xlabel('Age')
plt.ylabel('Spending Score')
plt.title('Age v/s Spending Score')
plt.show()
```

```python
se=[]
k_values=range(1,11)
from sklearn.cluster import KMeans
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df[['Age', 'Spending Score (1-100)']])
    se.append(kmeans.inertia_)
```

```python
se
```

```python
plt.plot(k_values,se,'bo-')
plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.title('kmeans clustering')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(df[['Age', 'Spending Score (1-100)']])
```

```python
labels = kmeans.labels_

plt.scatter(df['Age'],df['Spending Score (1-100)'],c=labels,cmap='Dark2')
plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.title('kmeans clustering')
plt.show()
```
