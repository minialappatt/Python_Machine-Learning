```python
import numpy as np

# Generate synthetic dataset
np.random.seed(0)
num_samples = 1500
num_features = 15
X = np.random.rand(num_samples, num_features)
y = np.random.rand(num_samples, 1)
```

```python
def vanilla_gradient_descent(X, y, learning_rate, num_iterations):
    num_samples, num_features = X.shape
    theta = np.zeros((num_features, 1))
    
    for _ in range(num_iterations):
        gradient = (1 / num_samples) * X.T.dot(X.dot(theta) - y)
        theta -= learning_rate * gradient
        
    return theta
```

```python
def mini_batch_gradient_descent(X, y, learning_rate, num_iterations, batch_size):
    num_samples, num_features = X.shape
    theta = np.zeros((num_features, 1))
    
    for _ in range(num_iterations):
        random_indices = np.random.choice(num_samples, batch_size, replace=False)
        X_batch = X[random_indices]
        y_batch = y[random_indices]
        
        gradient = (1 / batch_size) * X_batch.T.dot(X_batch.dot(theta) - y_batch)
        theta -= learning_rate * gradient
        
    return theta
```

```python
def stochastic_gradient_descent(X, y, learning_rate, num_iterations):
    num_samples, num_features = X.shape
    theta = np.zeros((num_features, 1))
    
    for _ in range(num_iterations):
        random_index = np.random.randint(num_samples)
        X_sample = X[random_index].reshape(1, -1)
        y_sample = y[random_index].reshape(1, -1)
        
        gradient = X_sample.T.dot(X_sample.dot(theta) - y_sample)
        theta -= learning_rate * gradient
        
    return theta
```

```python
import matplotlib.pyplot as plt

# Hyperparameters
learning_rate = 0.01
num_iterations = 100

# Try different batch sizes
batch_sizes = [1, 32, 128, 256]
losses = []

for batch_size in batch_sizes:
    theta = mini_batch_gradient_descent(X, y, learning_rate, num_iterations, batch_size)
    
    # Calculate loss
    loss = np.mean((X.dot(theta) - y) ** 2)
    losses.append(loss)

# Plot the convergence for different batch sizes
plt.figure(figsize=(10, 6))
plt.plot(batch_sizes, losses, marker='o')
plt.xlabel('Batch Size')
plt.ylabel('Loss')
plt.title('Convergence of Mini-Batch Gradient Descent')
plt.grid(True)
plt.show()
```
