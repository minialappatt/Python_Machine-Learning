```python
# Example of filling missing values with the mean
data_df.fillna(data_df.mean(), inplace=True)
# Example of dropping rows with missing values
housing_df.dropna(inplace=True)

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histograms
data_df.hist(figsize=(12, 8))
plt.show()

# Scatter plot for housing data if applicable
sns.scatterplot(x='longitude', y='latitude', data=housing_df, hue='median_house_value', alpha=0.5)
plt.show()

# Correlation matrix heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(data_df.corr(), annot=True, fmt=".2f")
plt.show()
```

```python
# Filter the data based on the conditions: more than 5 years of experience and salary over 60000
filtered_data = data_df[(data_df['YearsExperience'] > 5) & (data_df['Salary'] > 60000)]

# Count the number of employees meeting the criteria
count = filtered_data.shape[0]

print("Number of employees with more than 5 years of experience and earning more than $60,000:", count)
```

```python
# Filter the data based on the condition: salary between 50000 and 80000
filtered_data = data_df[(data_df['Salary'] >= 50000) & (data_df['Salary'] <= 80000)]

# Count the number of employees meeting the criteria
count = filtered_data.shape[0]

print("Number of employees earning between $50,000 and $80,000:", count)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import pandas as pd
import numpy as np

# Assuming 'data' DataFrame is already loaded with 'YearsExperience' and 'Salary'
X = data_df[['YearsExperience']]  # Predictor, needs to be reshaped or kept in DataFrame format
y = data_df['Salary']  # Response variable

# Splitting the data into training and testing sets (60:40 ratio)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

# Creating a linear regression model
model = LinearRegression()
model.fit(X_train, y_train)  # Training the model

# Making predictions on the test data
y_pred = model.predict(X_test)

# Calculating the r2_score
new_r2_score = r2_score(y_test, y_pred)
print("New r2_score for the model with a 60:40 training/testing split:", new_r2_score)

```
